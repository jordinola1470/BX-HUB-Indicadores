```python
import pandas as pd
```

```python
#DATA TRIMESTRAL
#Colombia
col_querie = pd.read_excel('Colombia/colombia_queries.xlsx',header=9)
col_catego = pd.read_excel('Colombia/colombia_categories.xlsx',header=9)

#Ecuador
ecu_querie = pd.read_excel('Ecuador/ecuador_queries.xlsx',header=9)
ecu_catego = pd.read_excel('Ecuador/ecuador_categories.xlsx',header=9)

#Peru
per_querie = pd.read_excel('Peru/peru_queries.xlsx',header=9)
per_catego = pd.read_excel('Peru/peru_categories.xlsx',header=9)
```

```python
#FUNCIONES

def estructura_querie(df_querie, df_catego):

    df_catego = df_catego.sort_values('weeks',ascending=False)    

    df = pd.concat([df_querie,df_catego])

    df.reset_index(inplace=True)
    df.drop(index=2,columns='index',inplace=True)
    df.set_index('weeks',inplace=True)

    for i in range(len(df.columns)):
        df.loc['Tasa Xenofobia',df.columns[i]] = (df.loc[df.index[1], df.columns[i]])/(df.loc[df.index[0], df.columns[i]])
    
    df_t = df.T
    df_t.reset_index(inplace=True)
    df_t['index'] = pd.to_datetime(df_t['index'])
    
    df_t['Mes'] = df_t['index'].dt.month
    df_t['A単o'] = df_t['index'].dt.year

    df_t.columns.name = None
    df_t = df_t.rename(columns={'index':'Fecha',df_t.columns[1]:'Volumen Total'})
    df_t = df_t[['Fecha','Mes','A単o','Volumen Total', 'Xenofobia', 'Tasa Xenofobia', ]] 

    return df_t
```

```python
#INDICADOR PAIS
colombia = estructura_querie(col_querie,col_catego)
ecuador = estructura_querie(ecu_querie,ecu_catego)
peru = estructura_querie(per_querie,per_catego)
```

```python
#EXPORTAR
with pd.ExcelWriter('Tasa Xenofobia-2023.xlsx') as writer:
    
    colombia.to_excel(writer,sheet_name='Colombia')
    ecuador.to_excel(writer,sheet_name='Ecuador')
    peru.to_excel(writer,sheet_name='Peru')
```
